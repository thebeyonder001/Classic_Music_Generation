```python
import requests
from bs4 import BeautifulSoup
```

```python
url='https://www.myntra.com/boys-tshirts/'
```

```python
req = requests.get(url)
```

```python
soup = BeautifulSoup(req.text, "lxml")
```

```python
print(soup.prettify())
```

```python
scripts = soup.find_all("script")
```

```python
i = 0
max = -1
for script in scripts:
    script_str = str(script)
    print(len(script_str))
    if len(str(scripts[max]))<len(script_str):
        max = i
    i = i+1
```

```python
#print(max)
max_script = str(scripts[max])
#print(max_script)
classes = max_script.split('products')
#print(classes[1])
temp = str(classes[1])
products = temp.split("sortOptions")
```

```python
print(len(products))
```

```python
urls = products[0].split('landingPageUrl')
urls = urls[1:]
```

```python
print(len(urls))
for url in urls:
    print(len(url))
```

```python
len(urls[0])
```

```python
len(urls[-1])
```

```python
imgs = []
for url in urls:
    string = url
    string = string.replace("u002F","")
    srcs=string.split("src")
    for src in srcs:
        src = src.replace(":","")
        src = src.replace("http","http:")
        a = src
        t = a.find(".jpg")
        s = list(a)
        s[t + 4:] = " "
        s = s[2:-1]
        a = ''.join(s)
        src = a
        imgs.append(src)
```

```python
for i in imgs:
    print(len(i))
```
